# 🎯 Objective of the day
The goal is to explore how LLMs improve their reasoning ability through two different scaling strategies:
1. Training-time scaling
2. Inference-time scaling

We will experiment with real models and observe how their answers change when we: increase the **model size** and/or increase the **reasoning effort** during inference.

The experiment consists of asking several logic puzzles to different models and comparing:
- correctness of the answer
- model size
- reasoning effort

## 🧪 Experiment Setup
We test different models on the same puzzle, choosing:
```text
model="gpt-5-nano" with reasoning_effort="minimal" and "low"
```
and
```text
model="gpt-5-mini" with reasoning_effort="minimal"
```
This allows us to observe two different scaling effects:
| Model      | Reasoning effort | Result      |
| ---------- | ---------------- | ----------- |
| gpt-5-nano | minimal          | often wrong |
| gpt-5-nano | low              | correct     |
| gpt-5-mini | minimal          | correct     |
