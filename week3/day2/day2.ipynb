# 🎯 Objective of the day
The goal is to explore how LLMs improve their reasoning ability through two different scaling strategies:
1. Training-time scaling
2. Inference-time scaling

We will experiment with real models and observe how their answers change when we: increase the **model size** and/or increase the **reasoning effort** during inference.

The experiment consists of asking several logic puzzles to different models and comparing:
- correctness of the answer
- model size
- reasoning effort

---

## 🧪 Experiment Setup
We test different models on the same puzzle, choosing:
```text
model="gpt-5-nano" with reasoning_effort="minimal" and "low"
```
and
```text
model="gpt-5-mini" with reasoning_effort="minimal"
```
This allows us to observe two different scaling effects:
| Model      | Reasoning effort | Result      |
| ---------- | ---------------- | ----------- |
| gpt-5-nano | minimal          | often wrong |
| gpt-5-nano | low              | correct     |
| gpt-5-mini | minimal          | correct     |

And this shows that both strategies can improve performance:
- increasing model size (training-time scaling)
- increasing reasoning effort (inference-time scaling)

---

# 🧩 Proposed Puzzles
To test this behavior we use three puzzles with increasing difficulty.

````python
easy_puzzle = [{"role": "user", "content": """You toss 2 coins. One of them is heads.
What is the probability the other is tails?
Answer with the probability only."""}]

medium_puzzle = [{"role": "user", "content": """You have two buckets: one holds 3 liters and the other holds 5 liters.
How can you measure exactly 4 liters?
Explain the steps."""}]

hard_puzzle = [{"role": "user", "content": """You are on a game show.
There are 3 doors:
- behind one door is a car
- behind the other two are goats.
You choose one door.
The host, who knows what is behind the doors, opens another door showing a goat.
You can either stay with your original choice or switch to the remaining door.
What strategy gives you the highest probability of winning the car?
Answer with the probability only."""}]
````

## ✅ Correct answers
Easy Puzzle:
```text
2/3
```

Medium Puzzle:
```text
1. Fill the 5L bucket
2. Pour into 3L bucket → 2L remain
3. Empty the 3L bucket
4. Pour the 2L into the 3L bucket
5. Fill the 5L bucket again
6. Pour into the 3L bucket until full → 4L remain in the 5L bucket
```

Hard Puzzle:
```text
Switch → 2/3
Stay → 1/3
```

````python
# imports

import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
````

````python
# key and models
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
openai = OpenAI()

MODEL_NANO = 'gpt-5-nano'
MODEL_MINI = 'gpt-5-mini'
````

````python
response_nano = openai.chat.completions.create(model=MODEL_NANO, messages=easy_puzzle, reasoning_effort="minimal")
display(Markdown(response_nano.choices[0].message.content))
````

MODEL_NANO and `reasoning_effort="minimal"` → wrong answer

````python
response_nano_2 = openai.chat.completions.create(model=MODEL_NANO, messages=easy_puzzle, reasoning_effort="low")
display(Markdown(response_nano_2.choices[0].message.content))
````

MODEL_NANO and `reasoning_effort="low"` → correct answer
