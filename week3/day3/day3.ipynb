# 🔀 Routers and Abstraction Layers for LLMs

In modern LLM applications, interacting with models is no longer limited to calling a single provider API.

Instead, developers often rely on tools that simplify access, improve flexibility, and enable dynamic model selection.

## 🎯 Objective

The goal of this notebook is to explore **three different ways to interact with LLMs** through higher-level tools:

1. [**OpenRouter**](https://openrouter.ai) → a router that allows access to multiple models through a single API key
2. [**LangChain**](https://www.langchain.com) → an abstraction layer for building structured LLM applications
3. [**LiteLLM**](https://www.litellm.ai) → a lightweight abstraction layer that standardizes API calls across providers

These tools introduce two key concepts:

- **Routers** → external services that forward your request to different providers
- **Abstraction layers** → local frameworks that unify how you call different models

---

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
say_hi = [{"role": "user", "content": "Hi!"}]
```

## OperRoute

```python
load_dotenv(override=True)
openrouter_api_key = os.getenv('OPENROUTER_API_KEY')

if openrouter_api_key:
    print(f"OpenRouter API Key exists and begins {openrouter_api_key[:3]}")
else:
    print("OpenRouter API Key not set (and this is optional)")
```

```python
openrouter_url = "https://openrouter.ai/api/v1"
openrouter = OpenAI(base_url=openrouter_url, api_key=openrouter_api_key)
```

```python
response = openrouter.chat.completions.create(model="openai/gpt-4o-mini", messages=say_hi)
display(Markdown(response.choices[0].message.content))
```

---

## Langchain

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini")
response = llm.invoke(say_hi)

display(Markdown(response.content))
```

---

## LiteLLM

```python
from litellm import completion
response = completion(model="openai/gpt-4o-mini", messages=say_hi)
reply = response.choices[0].message.content
display(Markdown(reply))

print(f"Input tokens: {response.usage.prompt_tokens}")
print(f"Output tokens: {response.usage.completion_tokens}")
print(f"Total tokens: {response.usage.total_tokens}")
print(f"Total cost: {response._hidden_params["response_cost"]*100:.4f} cents")
```
